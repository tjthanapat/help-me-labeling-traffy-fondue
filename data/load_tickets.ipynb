```python
import firebase_admin
from firebase_admin import credentials

cred = credentials.Certificate('serviceAccountKey.json')
firebase_admin.initialize_app(cred)
```

```python
from firebase_admin import firestore
db = firestore.client()
```

```python
for idx, row in my_tickets.iterrows():
    ticket_id = str(row['id']).zfill(5)
    ticket_key = idx
    ticket_content = row['content']
    doc_ref = db.collection(u'tickets').document(ticket_id)
    doc_ref.set({
        u'key': idx,
        u'content': ticket_content,
        u'isAssessed': False
    })
    print(f'Completed writing ticket with id {ticket_id}')
```
